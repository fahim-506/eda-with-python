#### np.corrcoef
- Return Pearson product-moment correlation coefficieants

```python
salary = np.array([20000,40000,25000,30000,50000])
experience = np.array([1,3,2,4,2])
np.corrcoef(salary,experience)
```

#### np.isin()
- checks whether each element of an array is present in another array


```python
x=np.array([1,2,3,4,5,6,76,7,43,1,45])
y=np.array([34,6,2,765,2,76,2,3,5,6,4])
x[np.isin(x,y)]
```

####  np.flip()

#### np.put()
- places values into an array at specific indices


```python
x[np.put(x,[1],[0])] # index 1 value 20 is updated to 0
```

#### np.delete()
- removes elements, rows, or columns from an array


```python
np.delete(x,[1,3])
```

### Set Functions


```python
m=np.array([1,2,3,4,5])
n=np.array([3,4,5,6,7])
np.union1d(m,n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n)
```

```python
np.setxor1d(m,n) #remove commen values
```

```python
# np.in1d(): checks whether each element of an array is present in another array
m[np.in1d(m,1)]
```

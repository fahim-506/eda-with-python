### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import pandas as pd
```

```python
#string
country = ['India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
runs = [13,24,56,78,100]
pd.Series(runs)
```

```python
marks = [50,67,99,87]
subjects = ['math','physics','cs','english']
pd.Series(marks,index=subjects)
```

```python
mark=pd.Series(marks,index=subjects,name='fff')
```

```python
mark
```

```python
mark = {
    'maths':67,
    'english':78,
    'hindi':100
}
mark_series=pd.Series(mark,name='fahim')
mark_series
```

## Series Attributes

```python
# size
mark_series.size
```

```python
# name
mark_series.name
```

```python
mark_series.is_unique
```

```python
mark_series.index
```

```python
mark_series.values
```

```python
# from a pandas dataframe
batsman = pd.read_csv('../datasets/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color="#CACD18")
plt.plot(batsman['index'],batsman['RG Sharma'],color="#15EF4F")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# solid , dotted , dashed , dashdot
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=1)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label = 'virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o', label = 'sharma')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc = 'upper center')        # display legend at the upper center
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)               # y-axis limits
plt.xlim(2017,2019)             # x-axis limits
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

# plt.show()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

# plt.show()
```

```python
plt.plot(
    batsman['index'],
    batsman['V Kohli'],
    label='Virat Kohli',
    color='#D9F10F',
    linestyle='solid',
    linewidth=3,
    marker='D',
    markersize=10,
    alpha=0.85
)

plt.plot(
    batsman['index'],
    batsman['RG Sharma'],
    label='Rohit Sharma',
    color='#FC00D6',
    linestyle='dashdot',
    linewidth=2,
    marker='o',
    alpha=0.85     #transparency
)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid(True)
plt.legend(loc='best')
plt.tight_layout()

plt.show()
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children, color='black')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('../datasets/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
np.arange(df.shape[0])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color = 'yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='green')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])        #display names in x axis
plt.show()
```

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

```python
# Convert categorical columns
cat_cols = [
    'First Name', 'Last Name', 'Country',
    'Car Brand', 'Car Model',
    'Car Color', 'Credit Card Type'
]

for col in cat_cols:
    df[col] = df[col].astype("category")

df.info()
```

```python
CURRENT_YEAR = 2025
df['Car Age'] = CURRENT_YEAR - df['Year of Manufacture']
```

```python
pd.DataFrame({'count': df.shape[0],  
              'cardinality': df.nunique()
             })
```

```python
pd.cut(df['Car Age'],bins=[0,5,10,20,30,50],labels=['05 yrs', '610 yrs', '1120 yrs', '2130 yrs', '30+ yrs'])
```

```python
sns.kdeplot(df['Car Age'], fill=True)
plt.title("Car Age Density Distribution")
plt.show()
```

```python
corr = df[['Year of Manufacture', 'Car Age']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

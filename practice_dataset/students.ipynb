```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('students.csv',index_col=False)
df
```

```python
df.info()
```

```python
df.describe(exclude='object')
```

```python
df.shape
```

```python
df.notnull().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df.info()
```

```python
df.dropna(ignore_index=True,inplace=True)
```

```python
df.drop_duplicates()
```

```python
df['Gender'].value_counts()
```

```python
# df[['Gender','EthnicGroup']].value_counts().unstack().unstack().unstack()
counts = df.groupby(['EthnicGroup', 'Gender']).size().unstack(fill_value=0)
counts
```

```python
# df[['Gender','EthnicGroup']].value_counts().plot.bar(color='grey')
counts.plot(kind='bar',color={'male': 'steelblue', 'female': 'salmon'})
plt.title('Count by Ethnic Group and Gender')
plt.ylabel('counts')
```

```python
df['Avg_Score'] = df[["MathScore", "ReadingScore", "WritingScore"]].mean(axis=1)
df['Avg_Score']
```

```python
plt.figure(figsize=(14,4))
for i, col in enumerate(["MathScore", "ReadingScore", "WritingScore"]):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"{col} Distribution")
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=df['MathScore'],kde=True)
plt.show()
```

```python
sns.histplot(data=df['ReadingScore'],kde=True)
plt.show()
```

```python
sns.histplot(data=df['WritingScore'],kde=True)
plt.show()
```

```python
sns.boxplot(x='Gender',y='Avg_Score',data=df)
plt.title("Average Score by Gender")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="TestPrep", y="Avg_Score", data=df)
plt.title("Impact of Test Preparation on Average Score")
plt.show()
```

```python


plt.figure(figsize=(10,5))
sns.boxplot(x="ParentEduc", y="Avg_Score", data=df)
plt.xticks(rotation=45)
plt.title("Average Score by Parent Education")
plt.show()
```

```python
plt.figure(figsize=(6,5))
corr = df[["MathScore", "ReadingScore", "WritingScore", "Avg_Score"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Scores")
plt.show()
```

```python
# np.where(condition, value_if_true, value_if_false)

df['Result'] = np.where (
    (df['Avg_Score']>60) & (df['TestPrep']=='completed') | (df['PracticeSport']=='regularly'),'pass' , 'fail'
    )

```

```python
df
```

```python
df['Result'].value_counts()
```

```python
counts = df['Result'].value_counts()
plt.bar(counts.index, counts.values, color=['green','red'])
plt.title('Number of Students: Pass vs Fail')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
```
